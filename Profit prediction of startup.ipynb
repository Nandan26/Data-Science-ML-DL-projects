```python
#import all th necessary library we needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#first we will load our data in data variable
#data.head will only print first five components of our data
data=pd.read_csv("50_Startups.csv")
data.head()
```

```python
#here we are finding profit so we are making it output and dropping it from input features
# here axis =1 means we are deleting entire data of profit if axis=0 then it will delete only one row
output=data['Profit']
Input_features= data.drop(['Profit'],axis=1)
```

```python
Input_features
```

```python
#Now State is a category but not a feature. ML algorithms see everything as numbers so we need to convert states into categories.
#we are making dummy variable for state category
dummy=pd.get_dummies(data['State'])
dummy
```

```python
#This looks good but has a huge problem of Dummy Variable Trap. In this we made three categories California, NewYork and Florida. But we just need two of them. For example we take California and Newyork, If there is a zero in both of them, this implies that the state to which startup belongs is Florida.
dummies=dummy.drop(['Florida'],axis=1)
dummies
```

```python
Input_features2 = Input_features.drop(['State'],axis = 1)
Input_features2
```

```python
#we will add our new columns (state) in our input features
Input_features2['Cal'] = dummies['California']
Input_features2['New Y'] = dummies['New York']
Input_features2
```

```python
#we will spilt our data into 80:20 ratio 80 for train and 20 for test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Input_features2, output, test_size = 0.2, random_state = 0)
```

```python
# we are treating this data as linear regression
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
# we are fitting the data in machine
linear.fit(X_train,y_train)
#we are predicting an output
test_pred=linear.predict(X_test)
test_pred
```

```python
# to plot dot graph
plt.scatter(Input_features2['R&D Spend'],output)
plt.show()
```

```python
# we have to find model accurasy
# 1. r square value
# 2. p value
# 3. getting the anknown coeeficient the betas
r_square=linear.score(X_test,y_test)
r_square
```

```python
#getting p value
from sklearn.feature_selection import f_regression
f = f_regression(X_test,y_test)
f
```

```python
#getting coef
coeff = linear.coef_
coeff
```

```python
#finding intersect beta 0
intercept = linear.intercept_
intercept
```

```python
# we also have one other library called states model for linear regression
import statsmodels.api as sm
X_train_new = sm.add_constant(X_train)#we can not find intercept using stats model so we are adding constant
X_train_new
```

```python
model = sm.OLS(y_train,X_train_new)
result = model.fit()
print(result.summary())
```
